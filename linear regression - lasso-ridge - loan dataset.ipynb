```python
import pandas as pd
```

```python
ld_train=pd.read_csv("loan_data_train.csv")
ld_test=pd.read_csv("loan_data_test.csv")
```

```python
import numpy as np
```

```python
#we need to combine train and test for data preparation
#so both train and test sets should have same set of columns
#test set does not have interest rate- so add a column interest rate with null values
ld_test["Interest.Rate"]=np.nan
```

```python
#once the sets are combined, we need to differentiate which is train data and which is test data
ld_train["data"]="train"
ld_test["data"]="test"
```

```python
#RHS=> checks if train columns are in the test, 
#if train columns is different, 
#when u give ld_test[ld_train.columns]=> it will throw error,cannot find column
#LHS=RHS => is done coz even if test has extra columns, test will be assigned with column 
#that are matched with train set only

ld_test=ld_test[ld_train.columns]
```

```python
#combining 2 sets
ld_all=pd.concat([ld_train,ld_test],axis=0)
```

```python
ld_all.head()
```

```python
ld_all.dtypes
```

```python
#dropping these column as they are not used in the prediction process
ld_all.drop(["ID","Amount.Funded.By.Investors"],axis=1,inplace=True)
```

```python
#these 2 columns have % at the end,we remove % so we can work on it 
for col in ["Interest.Rate","Debt.To.Income.Ratio"]:
    ld_all[col]=ld_all[col].str.replace("%","")
```

```python
ld_all.head()
```

```python
#these columns are objects - we need to convert them to numeric to use as x variables
#x variables can only be numeric
#errors=coerce => in interest.rate-there is null values(NA)-> when converted to numeric,throws coercion error
#cannot convert NA into numeric
#so if we give errors=coerce, it will ignore this error and compute

for col in ["Amount.Requested","Interest.Rate","Debt.To.Income.Ratio","Open.CREDIT.Lines","Revolving.CREDIT.Balance"]:
    ld_all[col]=pd.to_numeric(ld_all[col],errors="coerce")
```

```python
ld_all.dtypes
```

```python
#fico - range between 2 numbers -> need to convert it into 
k=ld_all["FICO.Range"].str.split("-",expand=True).astype(float)
ld_all["fico"]=0.5*(k[0]+k[1])

del ld_all["FICO.Range"]
```

```python
ld_all.head()
```

```python
ld_all["Employment.Length"].value_counts()
```

```python
ld_all["Employment.Length"]=ld_all["Employment.Length"].str.replace("years","")
ld_all["Employment.Length"]=ld_all["Employment.Length"].str.replace("year","")
ld_all["Employment.Length"]=np.where(ld_all["Employment.Length"].str[:2]=="10",10,ld_all["Employment.Length"])
ld_all["Employment.Length"]=np.where(ld_all["Employment.Length"].str[0]=="<1",0,ld_all["Employment.Length"])
#ld_all["Employment.Length"]=ld_all["Employment.Length"].str.replace("<1","1)
ld_all["Employment.Length"]=pd.to_numeric(ld_all["Employment.Length"],errors="coerce")
```

```python
ld_all.head()
```

```python
cat_col=ld_all.select_dtypes(["object"]).columns
cat_col
```

```python
cat_col=cat_col[:-1]
cat_col
```

```python
#dummification
for col in cat_col:
    freq=ld_all[col].value_counts()
    print(freq)
```

```python
for col in cat_col:
    freq=ld_all[col].value_counts()
    k=freq.index[freq>20][:-1]
    print(k)
```

```python
for col in cat_col:
    freq=ld_all[col].value_counts()
    k=freq.index[freq>20][:-1]
    for cat in k:
        name=col+'_'+cat
        ld_all[name]=(ld_all[col]==cat).astype(int)
    del ld_all[col]
```

```python
ld_all.isnull().sum()
```

```python
#eliminating null values
for col in ld_all.columns:
    if(col not in ['Interest.Rate','data'])&(ld_all[col].isnull().sum()>0):
        ld_all.loc[ld_all[col].isnull(),col]=ld_all.loc[ld_all['data']=='train',col].mean()
```

```python
ld_all.isnull().sum()
```

```python
ld_train=ld_all[ld_all["data"]=="train"]
del ld_train["data"]
ld_test=ld_all[ld_all["data"]=="test"]
ld_test.drop(["Interest.Rate","data"],axis=1,inplace=True)
```

```python
del ld_all
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting data
#since test data doesnot have interestrate(y), we cannot know if our model is working properly
#so we cannot use the test data directly
#so we split train data into 2- 
#train1-for building model(training set) , train2-for testing model
#if model works well with the train 2 set, it means the model accuracy is gud
#so with that we can apply the model to test data and predict interest rate

ld_train1,ld_train2=train_test_split(ld_train,test_size=0.2,random_state=2)
```

```python
x_train1=ld_train1.drop("Interest.Rate",axis=1)
y_train1=ld_train1["Interest.Rate"]
```

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV
```

```python
#penalty-100 values
#for each penalty value - compute coeff of insignificant variables and find model accuracy
lambdas=np.linspace(1,100,100)
```

```python
params={'alpha':lambdas}
```

```python
model=Ridge(fit_intercept=True)
```

```python
grid_search=GridSearchCV(model,param_grid=params,cv=10,scoring="neg_mean_absolute_error")
```
